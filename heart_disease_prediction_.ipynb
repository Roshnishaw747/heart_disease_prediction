```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (if not already loaded)
df = pd.read_csv("heart.csv")

# Count how many people of each age have disease = 1 and disease = 0
age_disease = df.groupby(['Age', 'HeartDisease']).size().unstack(fill_value=0)

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(age_disease.index, age_disease[0], label='No Disease', color='skyblue', linewidth=2)
plt.plot(age_disease.index, age_disease[1], label='Disease', color='salmon', linewidth=2)

# Add labels and title
plt.title("Line Chart: Heart Disease by Age")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
